```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(-25, 25):
    if i != 1:
        plt.scatter(i, 1, marker='_', color='blue')
    else:
        plt.scatter(i, 1, marker='*', color='red')
```

```python
lst = np.array(range(-100, 100))
```

```python
parabola = lst ** 2
```

```python
plt.scatter(lst, parabola, marker='.');
```

```python
new_parabola = np.where(lst == 50, 1, parabola )
```

```python
plt.scatter(lst, new_parabola, marker='.')
```

```python
# arbitrary calculation. Undefined at 3
def calc(x):
    undefined = 3
    if x == undefined:
        return 0
    else:
        return round((x ** 3 - 3*x**2)/(5*x-15), 3)
```

```python
left = np.linspace(-100, 100, 1000)
```

```python
left_calc = [calc(x) for x in left]
```

```python
vcalc = np.vectorize(calc)
```

```python
left_calc = vcalc(left)
```

```python
plt.plot(left, left_calc)
```

```python
calc(3)
```

An **asymptote** is a line or curve that approaches a given curve arbitrarily closely, as illustrated in the above diagram.
